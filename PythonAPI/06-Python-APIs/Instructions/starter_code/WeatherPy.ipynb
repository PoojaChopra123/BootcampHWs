```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key
cities_names = []
counter = 1
set_number = 1
print("Beginning Data Retrieval")
print("----------------------------")
for city in cities:
 
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + api_key + "&q=" + city
    weather_json = requests.get(url).json()
    
    if weather_json["cod"] == 200:
        city_name = weather_json["name"]
        cities_names.append(city_name)
        print(f"Processing Record {counter} of Set {set_number} | {city_name}")
        print(f"{url}")
        counter = counter + 1
        if counter == 50:
            set_number = set_number + 1
            counter = 0
            
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
#for city in cities:
temperature = []
humidity = []
cloudiness = []
wind_speed = []
latitude = []
longitude = []
date = []
country = []




for city in cities:
 
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + api_key + "&q=" + city
    weather_json = requests.get(url).json()
    if weather_json["cod"] == 200:
        
        temp = weather_json["main"]["temp"]
        hum = weather_json["main"]["humidity"] 
        cloud = weather_json["clouds"]["all"]
        ws = weather_json["wind"]["speed"]
        lat = weather_json["coord"]["lat"]
        lng = weather_json["coord"]["lon"]
        dt = weather_json["dt"]
        ct = weather_json["sys"]["country"]
        temperature.append(temp)
        humidity.append(hum)
        cloudiness.append(cloud)
        wind_speed.append(ws)
        latitude.append(lat)
        longitude.append(lat)
        date.append(dt)
        country.append(ct)
        
        

```

```python
length = {"City": len(cities_names),
         "Cloudiness": len(cloudiness),
         "Country": len(country),
         "Date": len(date),
         "Humidity": len(humidity),
         "Lat": len(latitude),
         "Lng": len(longitude),
         "Temp": len(temperature),
          "Wind Speed": len(wind_speed)}
length
```

```python

city_data = pd.DataFrame({"City": cities_names,
                          "Cloudiness": cloudiness,
                          "Country": country,
                          "Date": date,
                          "Humidity": humidity,
                          "Lat": latitude,
                          "Lng": longitude,
                          "Max Temp": temperature,
                          "Wind Speed": wind_speed})

city_data.head()
```
